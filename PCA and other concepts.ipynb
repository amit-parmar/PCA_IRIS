# PCA

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
df=pd.read_csv('iris.csv')
df.head()
```

```python
df['variety'].value_counts()
```

```python
target = df['variety']
```

```python
X = df.drop(["variety"], axis=1)
```

```python
X
```

```python
pca = PCA()
 
```

```python
X_tran = pca.fit_transform(X)
```

```python
X_tran
```

```python
pca.explained_variance_
```

```python
pca.explained_variance_ratio_
```

```python
pca = PCA(n_components=2)
```

```python
X_tran = pca.fit_transform(X)
```

```python
X_tran
```

```python
pca1=X_tran[:,0]
pca2=X_tran[:,1]
pca1
```

```python
color_dict={}
color_dict["Setosa"]="blue"
color_dict["Virginica"]="green"
color_dict["Versicolor"]="red"
```

```python
j=0
for i in target:
    print(i)
    plt.scatter(pca1[j], pca2[j], color=color_dict[i])
    j+=1
```
